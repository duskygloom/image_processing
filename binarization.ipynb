```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

image = Image.open("coffee.jpeg")
height = 400
width = int(image.width * height / image.height)
image = image.resize((width, height))
image = image.convert("L")

matrix = np.array(image)
hist = image.histogram()
hist_heigth = max(hist)

plt.plot(range(256), hist)
plt.show()

display(image)
```

## Static binarization with 2 divisions

```python
t1 = 256 / 2

plt.plot(np.linspace(t1, t1), np.linspace(0, hist_heigth), linestyle='--', color='skyblue')
plt.plot(range(256), hist)
plt.show()

edit = matrix.copy()
for i in range(height):
    for j in range(width):
        if edit[i, j] < t1:
            edit[i, j] = 0
        else:
            edit[i, j] = 255

display(Image.fromarray(edit))
```

## Static binarization with 4 divisions

```python
t2 = 256 / 2
t1 = t2 / 2
t3 = t2 + t1

plt.plot(np.linspace(t1, t1), np.linspace(0, hist_heigth), linestyle='--', color='skyblue')
plt.plot(np.linspace(t2, t2), np.linspace(0, hist_heigth), linestyle='--', color='skyblue')
plt.plot(np.linspace(t3, t3), np.linspace(0, hist_heigth), linestyle='--', color='skyblue')
plt.plot(range(256), hist)
plt.show()

edit = matrix.copy()
for i in range(height):
    for j in range(width):
        if edit[i, j] < t1:
            edit[i, j] = 0
        elif edit[i, j] < t2:
            edit[i, j] = 255 / 3
        elif edit[i, j] < t3:
            edit[i, j] = 2 * 255 / 3
        else:
            edit[i, j] = 255

display(Image.fromarray(edit))
```

## Mean binarization with 2 divisions

```python
t1 = np.mean(matrix.flatten())

plt.plot(np.linspace(t1, t1), np.linspace(0, hist_heigth), linestyle='--', color='skyblue')
plt.plot(range(256), hist)
plt.show()

edit = matrix.copy()

for i in range(height):
    for j in range(width):
        if edit[i, j] < t1:
            edit[i, j] = 0
        else:
            edit[i, j] = 255

display(Image.fromarray(edit))
```

## Mean binarization with 4 divisions

```python
t2 = np.mean(matrix.flatten())
t1 = np.mean([i for i in matrix.flatten() if i < t2])
t3 = np.mean([i for i in matrix.flatten() if i >= t2])

print(t1, t2, t3)

plt.plot(np.linspace(t1, t1), np.linspace(0, hist_heigth), linestyle='--', color='skyblue')
plt.plot(np.linspace(t2, t2), np.linspace(0, hist_heigth), linestyle='--', color='skyblue')
plt.plot(np.linspace(t3, t3), np.linspace(0, hist_heigth), linestyle='--', color='skyblue')
plt.plot(range(256), hist)
plt.show()

edit = matrix.copy()

for i in range(height):
    for j in range(width):
        if edit[i, j] < t1:
            edit[i, j] = 0
        elif edit[i, j] < t2:
            edit[i, j] = 255 / 3
        elif edit[i, j] < t3:
            edit[i, j] = 2 * 255 / 3
        else:
            edit[i, j] = 255

display(Image.fromarray(edit))
```
