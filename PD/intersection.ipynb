```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
height, width = 200, 200
```

```python
def get_binary_image(img: Image.Image) -> Image.Image:
    arr = np.array(img)
    mean = np.median(arr.flatten())
    for i in range(img.height):
        for j in range(img.width):
            arr[i, j] = 0 if arr[i, j] < mean else 255
    return Image.fromarray(arr, mode='L')
```

```python
anya = get_binary_image(Image.open('../assets/anya_meh.jpg').resize((width, height)).convert('L'))
anya_arr = np.array(anya)
raiden = get_binary_image(Image.open('../assets/raiden.jpeg').resize((width, height)).convert('L'))
raiden_arr = np.array(raiden)
```

```python
display(anya)
plt.plot(anya.histogram())
```

```python
display(raiden)
plt.plot(raiden.histogram())
```

```python
intersect_arr = np.zeros((height, width), np.uint8)
for i in range(height):
    for j in range(width):
        intersect_arr[i, j] = anya_arr[i, j] and raiden_arr[i, j]
```

```python
intersect_img = Image.fromarray(intersect_arr, mode='L')
```

```python
display(intersect_img)
plt.plot(intersect_img.histogram())
```
