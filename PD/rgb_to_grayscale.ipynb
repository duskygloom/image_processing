```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
height, width = 200, 200
```

```python
rgb = Image.open('../assets/fern_pout.webp').resize((width, height))
rgb_arr = np.array(rgb)
```

```python
display(rgb)
hist = rgb.histogram()
plt.plot(hist[:256], color='red')
plt.plot(hist[256:512], color='blue')
plt.plot(hist[512:], color='green')
```

```python
def rgb_to_gray(r: int, g: int, b: int) -> int:
    R = r / 255
    G = g / 255
    B = b / 255
    return round(255 * (0.299 * R + 0.587 * G + 0.114 * B))
```

```python
gray_arr = np.zeros((height, width), np.uint8)
for i in range(height):
    for j in range(width):
        gray_arr[i, j] = rgb_to_gray(*rgb_arr[i, j])
gray = Image.fromarray(gray_arr, 'L')
```

```python
display(gray)
plt.plot(gray.histogram())
```

```python
fig, axes = plt.subplots(2, 2)

axes[0][0].imshow(rgb)
hist = rgb.histogram()
axes[0][1].plot(hist[:256], color='red')
axes[0][1].plot(hist[256:512], color='green')
axes[0][1].plot(hist[512:], color='blue')
axes[1][0].imshow(gray, 'gray')
axes[1][1].plot(gray.histogram())
```
