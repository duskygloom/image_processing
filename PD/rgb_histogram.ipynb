```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
height, width = 200, 200
```

```python
img = Image.open('../assets/golden_retriever_babies.webp').resize((height, width))
arr = np.array(img)
```

```python
display(img)
```

```python
hist = img.histogram()
r = hist[:256]
g = hist[256:512]
b = hist[512:]
```

```python
display(plt.plot(r, color='red'))
display(plt.plot(g, color='green'))
display(plt.plot(b, color='blue'))
```

```python
arr_r = np.zeros((height, width, 3), np.uint8)
arr_g = np.zeros((height, width, 3), np.uint8)
arr_b = np.zeros((height, width, 3), np.uint8)
```

```python
for i in range(height):
    for j in range(width):
        arr_r[i, j, 0] = arr[i, j, 0]
        arr_g[i, j, 1] = arr[i, j, 1]
        arr_b[i, j, 2] = arr[i, j, 2]
```

```python
img_r = Image.fromarray(arr_r)
img_g = Image.fromarray(arr_g)
img_b = Image.fromarray(arr_b)
```

```python
display(img_r)
plt.plot(img_r.histogram()[:256])
```

```python
display(img_g)
plt.plot(img_g.histogram()[256:512])
```

```python
display(img_b)
plt.plot(img_b.histogram()[512:])
```
