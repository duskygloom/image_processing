```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
height, width = 200, 200
```

```python
img = Image.open('../assets/shiba_inu.jpeg').convert('L').resize((width, height))
arr = np.array(img)
```

```python
display(img)
plt.plot(img.histogram())
```

```python
mirror_arr = arr.copy()
for i in range(height):
    mirror_arr[i] = mirror_arr[i][::-1]
mirror_img = Image.fromarray(mirror_arr, 'L')
```

```python
display(mirror_img)
plt.plot(mirror_img.histogram())
```

```python
flip_arr = arr.copy()
for i in range(height):
    flip_arr[i] = arr[height-i-1]
flip_img = Image.fromarray(flip_arr, 'L')
```

```python
display(flip_img)
plt.plot(flip_img.histogram())
```

```python
fig, axes = plt.subplots(2, 3)
axes[0, 0].imshow(img, 'gray')
axes[0, 1].imshow(mirror_img, 'gray')
axes[0, 2].imshow(flip_img, 'gray')
axes[1, 0].plot(img.histogram())
axes[1, 1].plot(mirror_img.histogram())
axes[1, 2].plot(flip_img.histogram())
```
