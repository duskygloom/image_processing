```python
import numpy as np
from PIL import Image
from PIL import ImageDraw
from PIL import ImageFont
import matplotlib.pyplot as plt
```

```python
height, width = 200, 200
name = 'Fern'
font_size = 20
```

```python
img = Image.open('../assets/fern_pout.webp').convert('L').resize((width, height))
arr = np.array(img)
```

```python
fig, axes = plt.subplots(1, 2)
axes[0].imshow(img, 'gray')
axes[1].plot(img.histogram())
```

```python
name_img = Image.new('L', (width, height), 0)
pen = ImageDraw.Draw(name_img, 'L')
pen.text((width-(len(name)*(font_size//2)), height-font_size*1.5), name, 255, ImageFont.truetype('../assets/Chewy-Regular.ttf', font_size))
name_arr = np.array(name_img)
```

```python
display(name_img)
```

```python
for i in range(height):
    for j in range(width):
        s = int(arr[i, j]) + int(name_arr[i, j])
        if s > 255:
            s = 255
        arr[i, j] = s
```

```python
display(Image.fromarray(arr))
```
