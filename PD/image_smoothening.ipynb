```python
import statistics
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
height, width = 300, 200

n_degree = 2
```

```python
img = Image.open('../assets/butterfly.png').convert('L').resize((width, height))
arr = np.array(img)
```

```python
fig, axes = plt.subplots(1, 2)
axes[0].imshow(img, 'gray')
axes[1].plot(img.histogram())
```

### **Average smoothing**

```python
mean_arr = arr.copy()

for i in range(height):
    for j in range(width):
        # total, count = 0, 0
        neighbours = []
        for k in range(-n_degree, n_degree+1):
            for l in range(-n_degree, n_degree+1):
                if i+k >= 0 and i+k < height and j+l >= 0 and j+l < width:
                    # count += 1
                    # total += int(arr[i+k, j+l])
                    neighbours.append(int(arr[i+k, j+l]))
        # arr[i, j] = int(total / count)
        mean_arr[i, j] = np.uint8(statistics.mean(neighbours))
```

```python
mean_image = Image.fromarray(mean_arr)
```

```python
fig, axes = plt.subplots(1, 2)
axes[0].set_title('Original')
axes[0].imshow(img, 'gray')
axes[1].set_title('Average')
axes[1].imshow(mean_image, 'gray')
```

### **Median smoothing**

```python
median_arr = arr.copy()

for i in range(height):
    for j in range(width):
        # total, count = 0, 0
        neighbours = []
        for k in range(-n_degree, n_degree+1):
            for l in range(-n_degree, n_degree+1):
                if i+k >= 0 and i+k < height and j+l >= 0 and j+l < width:
                    # count += 1
                    # total += int(arr[i+k, j+l])
                    neighbours.append(int(arr[i+k, j+l]))
        # arr[i, j] = int(total / count)
        median_arr[i, j] = np.uint8(statistics.median(neighbours))
```

```python
median_image = Image.fromarray(median_arr)
```

```python
fig, axes = plt.subplots(1, 3)
axes[0].set_title('Original')
axes[0].imshow(img, 'gray')
axes[1].set_title('Average')
axes[1].imshow(mean_image, 'gray')
axes[2].set_title('Median')
axes[2].imshow(median_image, 'gray')
```
