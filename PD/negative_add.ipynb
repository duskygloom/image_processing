```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
height, width = 200, 200
```

```python
coffee = Image.open('assets/coffee.jpeg').convert('L').resize((height, width))
coffee_data = np.array(coffee)
```

### **Image A**

```python
display(coffee)
display(plt.plot(coffee.histogram()))
```

```python
landscape = Image.open('assets/landscape.jpg').convert('L').resize((height, width))
landscape_data = np.array(landscape)
```

### **Image B**

```python
display(landscape)
display(plt.plot(landscape.histogram()))
```

```python
def get_image_sum(a: Image.Image, b: Image.Image) -> Image.Image:
    a_arr = np.array(a)
    b_arr = np.array(b)
    for i in range(height):
        for j in range(width):
            s = int(a_arr[i, j]) + int(b_arr[i, j])
            if s > 255:
                s = 255
            a_arr[i, j] = s
    c = a.copy()
    c.putdata(a_arr.flatten())
    return c
```

```python
sum_image = Image.new('L', (height, width))

sum_array = np.empty((height, width))

sum_image = get_image_sum(coffee, landscape)
```

### **Image A + Image B**

```python
display(sum_image)
display(plt.plot(sum_image.histogram()))
```

```python
def get_negative_image(img: Image.Image) -> Image.Image:
    arr = np.array(img)
    arr = arr.flatten()
    for i in range(len(arr)):
        arr[i] = 255 - arr[i]
    newimg = img.copy()
    newimg.putdata(arr)
    return newimg
```

### **Negative A**

```python
negative_coffee = get_negative_image(coffee)

display(negative_coffee)
display(plt.plot(negative_coffee.histogram()))
```

### **Negative B**

```python
negative_landscape = get_negative_image(landscape)

display(negative_landscape)
display(plt.plot(negative_landscape.histogram()))
```

```python
negative_sum_image = Image.new('L', (height, width))

negative_sum_array = np.empty((height, width))

negative_sum_image = get_image_sum(negative_coffee, negative_landscape)
```

### **Negative A + Negative B**

```python
display(negative_sum_image)
display(plt.plot(negative_sum_image.histogram()))
```
