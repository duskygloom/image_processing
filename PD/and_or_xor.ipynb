```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
height, width = 200, 200
```

```python
def get_binary_image(img: Image.Image) -> Image.Image:
    arr = np.array(img)
    mean = np.median(arr.flatten())
    for i in range(img.height):
        for j in range(img.width):
            arr[i, j] = 0 if arr[i, j] < mean else 255
    return Image.fromarray(arr, mode='L')
```

```python
anya = get_binary_image(Image.open('../assets/anya_meh.jpg').resize((width, height)).convert('L'))
anya_arr = np.array(anya)
raiden = get_binary_image(Image.open('../assets/raiden.jpeg').resize((width, height)).convert('L'))
raiden_arr = np.array(raiden)
```

```python
fig, axes = plt.subplots(2, 2)
axes[0, 0].imshow(anya, 'gray')
axes[0, 1].plot(anya.histogram())
axes[1, 0].imshow(raiden, 'gray')
axes[1, 1].plot(raiden.histogram())
```

```python
and_arr = np.zeros((height, width), np.uint8)
for i in range(height):
    for j in range(width):
        and_arr[i, j] = anya_arr[i, j] and raiden_arr[i, j]
```

```python
and_img = Image.fromarray(and_arr, 'L')
```

```python
fig, axes = plt.subplots(1, 3)
axes[0].imshow(anya, 'gray')
axes[0].set_title('Image A')
axes[1].imshow(raiden, 'gray')
axes[1].set_title('Image B')
axes[2].imshow(and_img, 'gray')
axes[2].set_title('A AND B')
```

```python
or_arr = np.zeros((height, width), np.uint8)
for i in range(height):
    for j in range(width):
        or_arr[i, j] = anya_arr[i, j] or raiden_arr[i, j]
```

```python
or_img = Image.fromarray(or_arr, 'L')
```

```python
fig, axes = plt.subplots(1, 3)
axes[0].imshow(anya, 'gray')
axes[0].set_title('Image A')
axes[1].imshow(raiden, 'gray')
axes[1].set_title('Image B')
axes[2].imshow(or_img, 'gray')
axes[2].set_title('A OR B')
```

```python
xor_arr = np.zeros((height, width), np.uint8)
for i in range(height):
    for j in range(width):
        xor_arr[i, j] = anya_arr[i, j] ^ raiden_arr[i, j]
```

```python
xor_img = Image.fromarray(xor_arr, 'L')
```

```python
fig, axes = plt.subplots(1, 3)
axes[0].imshow(anya, 'gray')
axes[0].set_title('Image A')
axes[1].imshow(raiden, 'gray')
axes[1].set_title('Image B')
axes[2].imshow(xor_img, 'gray')
axes[2].set_title('A XOR B')
```
