```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
height, width = 200, 200
```

```python
obj = Image.open('../assets/chill_guy.png').resize((width, height))
obj_arr = np.array(obj)
```

```python
display(obj)
hist = obj.histogram()
plt.plot(hist[:256], color='red')
plt.plot(hist[256:512], color='green')
plt.plot(hist[512:768], color='blue')
plt.plot(hist[768:], color='grey')
```

```python
scene = Image.open('../assets/landscape.jpeg').resize((int(width*1.5), height)).convert('L')
scene_arr = np.array(scene)
```

```python
display(scene)
plt.plot(scene.histogram())
```

```python
def rgb_to_gray(r: int, g: int, b: int) -> int:
    R = r / 255
    G = g / 255
    B = b / 255
    return round(255 * (0.299 * R + 0.587 * G + 0.114 * B))
```

```python
for i in range(height):
    for j in range(width):
        if obj_arr[i, j, 3] > 0:
            scene_arr[i, j] = rgb_to_gray(*obj_arr[i, j][:3])
```

```python
scene = Image.fromarray(scene_arr)
```

```python
display(scene)
plt.plot(scene.histogram())
```
