```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
height, width = 200, 200
```

```python
anya = Image.open('../assets/anya_meh.jpg').resize((width, height)).convert('L')
anya_arr = np.array(anya)
raiden = Image.open('../assets/raiden.jpeg').convert('L').resize((width, height))
raiden_arr = np.array(raiden)
```

```python
fig, axes = plt.subplots(2, 2)
axes[0, 0].imshow(anya, 'gray')
axes[0, 1].plot(anya.histogram())
axes[1, 0].imshow(raiden, 'gray')
axes[1, 1].plot(raiden.histogram())
```

```python
sum_arr = np.zeros((height, width), np.uint8)
for i in range(height):
    for j in range(width):
        s = int(anya_arr[i, j]) + int(raiden_arr[i, j])
        if s > 255:
            s = 255
        sum_arr[i, j] = s
```

```python
sum_img = Image.fromarray(sum_arr, 'L')
```

```python
fig, axes = plt.subplots(2, 3)
axes[0, 0].imshow(anya, 'gray')
axes[0, 0].set_title('Image A')
axes[0, 1].imshow(raiden, 'gray')
axes[0, 1].set_title('Image B')
axes[0, 2].imshow(sum_img, 'gray')
axes[0, 2].set_title('A + B')
axes[1, 0].plot(anya.histogram())
axes[1, 1].plot(raiden.histogram())
axes[1, 2].plot(sum_img.histogram())
```

```python
sub_arr = np.zeros((height, width), np.uint8)
for i in range(height):
    for j in range(width):
        s = int(anya_arr[i, j]) - int(raiden_arr[i, j])
        if s < 0:
            s = 0
        sub_arr[i, j] = s
```

```python
sub_img = Image.fromarray(sub_arr, 'L')
```

```python
fig, axes = plt.subplots(2, 3)
axes[0, 0].imshow(anya, 'gray')
axes[0, 0].set_title('Image A')
axes[0, 1].imshow(raiden, 'gray')
axes[0, 1].set_title('Image B')
axes[0, 2].imshow(sub_arr, 'gray')
axes[0, 2].set_title('A - B')
axes[1, 0].plot(anya.histogram())
axes[1, 1].plot(raiden.histogram())
axes[1, 2].plot(sub_img.histogram())
```

```python
sub_arr = np.zeros((height, width), np.uint8)
for i in range(height):
    for j in range(width):
        s = int(raiden_arr[i, j]) - int(anya_arr[i, j])
        if s < 0:
            s = 0
        sub_arr[i, j] = s
```

```python
sub_img = Image.fromarray(sub_arr, 'L')
```

```python
fig, axes = plt.subplots(2, 3)
axes[0, 0].imshow(anya, 'gray')
axes[0, 0].set_title('Image A')
axes[0, 1].imshow(raiden, 'gray')
axes[0, 1].set_title('Image B')
axes[0, 2].imshow(sub_img, 'gray')
axes[0, 2].set_title('B - A')
axes[1, 0].plot(anya.histogram())
axes[1, 1].plot(raiden.histogram())
axes[1, 2].plot(sub_img.histogram())
```
