```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
rgb = Image.open('../assets/raiden.jpeg')
```

```python
fig, axes = plt.subplots(1, 2)
axes[0].imshow(rgb)
hist = rgb.histogram()
axes[1].plot(hist[:256], color='red')
axes[1].plot(hist[256:512], color='green')
axes[1].plot(hist[512:], color='blue')
```

```python
black_white = rgb.convert('L')
```

```python
fig, axes = plt.subplots(1, 2)
axes[0].imshow(black_white, 'gray')
axes[1].plot(black_white.histogram())
```

```python
new_size = black_white.resize((200, 200))
```

```python
fig, axes = plt.subplots(1, 2)
axes[0].imshow(new_size, 'gray')
axes[1].plot(new_size.histogram())
```

```python
fig, axes = plt.subplots(1, 3)
axes[0].imshow(rgb)
axes[0].set_title('Original')
axes[1].imshow(black_white, 'gray')
axes[1].set_title('Black and white')
axes[2].imshow(new_size, 'gray')
axes[2].set_title('Resized')
```
