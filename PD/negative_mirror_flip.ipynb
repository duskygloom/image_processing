```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
height, width = 200, 200
```

```python
img = Image.open('../assets/shiba_inu.jpeg').convert('L').resize((width, height))
arr = np.array(img)
```

```python
fig, axes = plt.subplots(1, 2)
axes[0].imshow(img, 'gray')
axes[1].plot(img.histogram())
```

```python
negative_arr = arr.copy()
for i in range(height):
    for j in range(width):
        negative_arr[i, j] = 255 - negative_arr[i, j]
negative_img = Image.fromarray(negative_arr)
```

```python
fig, axes = plt.subplots(2, 2)
axes[0, 0].set_title('Original')
axes[0, 0].imshow(img, 'gray')
axes[1, 0].plot(img.histogram())
axes[0, 1].set_title('Negative')
axes[0, 1].imshow(negative_img, 'gray')
axes[1, 1].plot(negative_img.histogram())
```

```python
mirror_arr = negative_arr.copy()
for i in range(height):
    mirror_arr[i] = mirror_arr[i][::-1]
mirror_img = Image.fromarray(mirror_arr, 'L')
```

```python
fig, axes = plt.subplots(2, 3)
axes[0, 0].set_title('Original')
axes[0, 0].imshow(img, 'gray')
axes[1, 0].plot(img.histogram())
axes[0, 1].set_title('Negative')
axes[0, 1].imshow(negative_img, 'gray')
axes[1, 1].plot(negative_img.histogram())
axes[0, 2].set_title('Mirror')
axes[0, 2].imshow(mirror_img, 'gray')
axes[1, 2].plot(mirror_img.histogram())
```

```python
flip_arr = negative_arr.copy()
for i in range(height):
    flip_arr[i] = negative_arr[height-i-1]
flip_img = Image.fromarray(flip_arr, 'L')
```

```python
fig, axes = plt.subplots(2, 4)
axes[0, 0].set_title('Original')
axes[0, 0].imshow(img, 'gray')
axes[1, 0].plot(img.histogram())
axes[0, 1].set_title('Negative')
axes[0, 1].imshow(negative_img, 'gray')
axes[1, 1].plot(negative_img.histogram())
axes[0, 2].set_title('Mirror')
axes[0, 2].imshow(mirror_img, 'gray')
axes[1, 2].plot(mirror_img.histogram())
axes[0, 3].set_title('Flip')
axes[0, 3].imshow(flip_img, 'gray')
axes[1, 3].plot(flip_img.histogram())
```
