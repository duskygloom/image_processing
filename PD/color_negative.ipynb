```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
height, width = 200, 200
```

```python
img = Image.open('../assets/smiling_cat.jpg').resize((width, height))
arr = np.array(img)
```

```python
display(img)
hist = img.histogram()
plt.plot(hist[:256], color='red')
plt.plot(hist[256:512], color='green')
plt.plot(hist[512:], color='blue')
```

```python
for i in range(height):
    for j in range(width):
        arr[i, j] = [255 - arr[i, j, k] for k in range(3)]
```

```python
negative_img = Image.fromarray(arr)
```

```python
display(negative_img)
hist = negative_img.histogram()
plt.plot(hist[:256], color='red')
plt.plot(hist[256:512], color='green')
plt.plot(hist[512:], color='blue')
```
