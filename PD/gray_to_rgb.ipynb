```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
height, width = 200, 200
```

```python
img = Image.open('../assets/cameraman.png').resize((width, height)).convert('L')
arr = np.array(img)
```

```python
color = np.zeros((width, height, 3), np.uint8)

for i in range(height):
    for j in range(width):
        color[i, j] = np.array([0.33*int(arr[i, j]), int(arr[i, j]), 0.11*int(arr[i, j])])
```

```python
color_img = Image.fromarray(color, 'RGB')
```
