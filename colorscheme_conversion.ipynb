```python
from PIL import Image

img = Image.open('assets/anya_meh.jpg')
height = 250
width = int(img.width * height / img.height)
img = img.resize((width, height))
display(img)
```

```python
import numpy as np

matrix = np.array(img)
```

```python
import random

random_color = random.choice(random.choice(matrix))
```

```python
print(random_color)
```

```python
random_color_matrix = np.array([[random_color for i in range(width)] for j in range(height)])
random_color_image = Image.fromarray(random_color_matrix)
display(random_color_image)
```

```python
total_color = 0
for i in random_color:
    total_color += int(i)
print(total_color)
```

```python
r = random_color[0] / total_color
g = random_color[1] / total_color
b = random_color[2] / total_color

print(r, g, b)
print(r + g + b)
```

```python
r = random_color[0] / 255
g = random_color[1] / 255
b = random_color[2] / 255

print(r, g, b)
```

```python
i = sum((r, g, b)) / 3
print(i)
```

```python
if i == 0:
    s = 0
else:
    s = 1 - min((r, g, b)) / i
print(s)
```

```python
numerator = ((r - g) + (r - b)) / 2
print(numerator)
```

```python
denominator = np.sqrt((r - g) * (r - g) + (r - b) * (g - b))
print(denominator)
```

```python
theta = np.rad2deg(np.acos(numerator / denominator))
print(theta)
```

```python
if b > g:
    h = 360 - theta
else:
    h = theta
```
